```python
import numpy as np
from utils import cluster_model, filter_model
 
np.random.seed(0)
embeddings = np.concatenate([
    np.random.normal(loc=[0, 0], scale=0.2, size=(100, 2)),
    np.random.normal(loc=[1, 1], scale=0.2, size=(100, 2)),
    np.random.normal(loc=[2, 2], scale=0.2, size=(100, 2)),
]) 

cm = cluster_model(num_clusters = 3)
clusters = cm.fit(embeddings)
fm = filter_model()
reps = embeddings[fm.fit(clusters)]

import matplotlib.pyplot as plt
plt.scatter(embeddings[:, 0], embeddings[:, 1], c='blue', alpha=0.5, label='Data Points')
plt.scatter(reps[:, 0], reps[:, 1], c='red', marker='x', s=100, label='Representatives')
plt.title('Data Points and Representatives')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.show()
```
